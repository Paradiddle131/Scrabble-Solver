```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('game_screen2.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=cmap)
```

```python
display(img)
```

```python
display(gray, cmap='gray')
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)[45:435]
display(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)[45:435]
display(edges)
```

```python
len(edges)
```

```python
edges.shape
```

```python
med_val = np.median(img)
lower = int(max(0, 0.7*med_val))
upper = int(max(255, 1.3*med_val))
```

```python
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper)[45:435]
display(edges, cmap='gray')
```
